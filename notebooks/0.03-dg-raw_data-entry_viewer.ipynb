```python
import pandas as pd

df_raw = pd.read_csv("../data/raw/haunted_places.tab", sep = "\t") 
df_cleaned = pd.read_csv("../data/processed/haunted_places_cleaned.tab", sep = "\t")
entry_no = 1413
```

```python
df_raw[df_raw.index == entry_no]
```

### Look for a word

```python
# Interesting keywords: {"indian: 258", "john: 69", "president": 28}
keyword = 'i'
df_raw[df_raw['description'].
        str.lower()
        .str.split()
        .explode()
        .str.strip()
        .eq(keyword)
        .groupby(level=0).any()
]

df_cleaned[df_cleaned['description'].
       str.lower()
       .str.split()
       .explode()
       .str.strip()
       .eq(keyword)
       .groupby(level=0).any()
]

```
